```python
import pandas as pd
import os
import matplotlib.pyplot as plt
#plt.style.reload_library()
```

```python
slo = {
    'las' :[],
    'pole':[],
    'pole uprawne':[],
    'woda':[],
    'zabudowa':[]
}
list_dict = list(slo.keys())
i = 0
b = 0
added = list_dict[0]
folder = os.walk(r'C:\aa_sem_i\CPO\Temat_1\dane_przetworzone\2_4b')
for x,y,files in folder:
    for file in files:
        file_size = os.stat(f'C:/aa_sem_i/CPO/Temat_1/dane_przetworzone/2_4b/{file}').st_size
        #print(file, file_size)
        slo[added].append(file_size)
        i += 1
        if i % 5 == 0:
            b += 1
            if b<5:
                added = list_dict[b]
        #print(i)
print(slo)
```

```python
quality = ['100','90','80','70','60']
plt.style.use("fast")
fig = plt.figure(figsize = (10,6), facecolor='white', dpi=200)
plt.title('Wykres zależności rozmiaru pliku od współczynnika obrazu', fontsize = 12, pad = 20)
plt.plot(quality, slo['zabudowa'],'D-', linewidth = 1.5)
plt.plot(quality, slo['pole uprawne'],'D-', linewidth = 1.5)
plt.plot(quality, slo['las'],'D-', linewidth = 1.5, c = 'brown' )
plt.plot(quality, slo['pole'],'D-', linewidth = 1.5)
plt.plot(quality, slo['woda'],'D-', linewidth = 1.5)
#plt.xticks(quality)
plt.legend(['zabudowa','pole uprawne','las','łąka/pastwisko','woda'])
plt.ylabel('Rozmiar pliku [byte]', fontsize = 12)
plt.xlabel('Współczynnik jakośc obrazu', fontsize = 12)
plt.grid(True)
plt.savefig(r'C:\aa_sem_i\CPO\Temat_1\dane_przetworzone\wykres.png')
```
