```python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt
from numpy import arange, array, transpose, linalg, polyfit, polyval
# import mplcursors
from matplotlib.ticker import MultipleLocator
import os
import numpy as np
os.chdir(r'C:\a_sem_i\PiPG_I\KOLOKWIUM')

S = np.loadtxt("0008b-19.txt")

t = arange(0, len(S))                       #wektor przechowujący epoki pomiarowe

ones_vector = []
for i in range(0, len(S)):                  #stworzenie wektora z jedynkami
   ones_vector.append(1)
#print(ones_vector)
#print(len(ones_vector))

A = list(zip(t, ones_vector))               #stworzenie macierzy współrzynników A
#print(array(A))'


#L = 40                                      #zdefiniowanie wielkości operatora przesunięcia tzw. lag
L = int(input("Wprowadź parametr operatora przesunięcia - tzw. lag (jako liczba całkowita): "))


delta_S = []
for i in range(0,len(S)-L):                 #operator różnicowy
    delta = S[i+L]-S[i]
    delta_S.append(delta)
#print(delta_S)

#x = linalg.inv((transpose(A) @ A)) @ transpose(A) @ S   #wyznaczenie parametrów prostej metodą najmniejszych kwadratów
#print(x)

#albo

x = polyfit(t, S, 1)                        #wyznaczenie parametrów prostej metodą najmniejszych kwadratów
#print(x)

V = A @ x - S                               #wektor odchyłek
#print(V)


plt.style.use('seaborn-bright')             #ustawienie stylu wykresu

plt.figure(figsize=(14, 9))

plt.subplot(311)
plt.title("Wykres wartości sygnału w zależności od epoki")
plt.plot(t, S)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(312)
t2 =t[:-L ]
plt.title("Wykres przetworzonego sygnału poprzez zastosowanie operatora różnicowego L = %i" %L)
plt.scatter(t2, delta_S, s=1)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(313)
plt.title("Wykres odchyłek wartości sygnału od wartości wpasowanej prostej")
plt.scatter(t, V, s=1)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)


plt.show()

#jump_epoch = [0, 208, 333, 613, 999]       #wymuszenie na użytkowniku wskazania epok skoków
jump_epoch = input("Na podstawie wykresów podaj epoki skoków (jako liczby całkowite rosnąco po znaku spacji): ")
jump_epoch = jump_epoch.split()
jump_epoch = list(map(int, jump_epoch))
odp1 = list(map(int, jump_epoch))
print("Epoki w których nastąpił skok : %r" %jump_epoch)
jump_epoch.insert(0, 0)                     #dodanie epoki początkowej do listy jump_epoch
jump_epoch.append(len(V)-1)                 #dodanie epoki końcowej do listy jump_epoch
#print(jump_epoch)
params = []
jumps = []

for i in range(0, len(jump_epoch)-1):       #wyznaczenie parametrów prostych o początku/końcu zdefiniowanym przez epoki wskazane przez użytkownika
    params.append(polyfit(t[jump_epoch[i]:jump_epoch[i+1]], V[jump_epoch[i]:jump_epoch[i+1]], 1))
#print(params)

#r = params[0][0]*t[:331] + params[0][1]
#print(r)

for i in range(1, len(jump_epoch)-1):       #wyznaczenie wartości skoków poprzez odjęcie wartości funkcji liniowych na końcach przedziałów
    q = jump_epoch[i]*params[i][0]+params[i][1] - ((jump_epoch[i]-1)*params[i-1][0]+params[i-1][1])
    #print("Koniec: ", (jump_epoch[i]*params[i-1][0]+params[i-1][1]))
    #print("Początek: ", jump_epoch[i] * params[i][0] + params[i][1])
    jumps.append(q)
print("Wielkości skoków: %r" %jumps)
odp2 = jumps

plt.figure(figsize=(14, 11))

plt.subplot(411)
plt.title("Wykres wartości sygnału w zależności od epoki")
plt.plot(t, S)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(412)
plt.title("Wykres przetworzonego sygnału poprzez zastosowanie operatora różnicowego L = %i" %L)
t2 =t[:-L ]
plt.scatter(t2, delta_S, s=1)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(413)
plt.title("Wykres odchyłek wartości sygnału od wartości wpasowanej prostej")
plt.scatter(t, V, s=1)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(414)
plt.title("Proste wpasowane w odchyłki")
plt.scatter(t, V, s=1)
for i in range(0,len(params)):
    P = params[i][0]*t + params[i][1]
    plt.plot(t, P)
plt.xticks(arange(0, len(S)+100, step=100))
plt.grid(b=True, which='major', color='#777777', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show()

f = open("Andrzej_Bielak_wyniki.txt", "w")              #zapis odpowiedzi do pliku tekstowego
f.write("PPG1 - Wyniki kolokwium    i = 1\n\n")
f.write("Epoka skoku    -     wielkosc skoku\n")
for i in range(0,len(jumps)):
    f.write("%r    -    %r\n" %(odp1[i], odp2[i]))
f.close()
```
