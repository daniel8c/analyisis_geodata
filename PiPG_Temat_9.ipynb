```python
import matplotlib.pyplot as plt
from numpy import array, sin, pi, linspace, append, random, arange, histogram, cos, fft, mean
import numpy as np
import scipy
import os
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
model = LinearRegression()

os.chdir(r'C:\a_sem_i\PiPG_I\temat_9')

# In[353]:
x,y = np.loadtxt('fft_data.txt',unpack = True)
x,y = x.reshape((-1,1)), y.reshape((-1,1))
                                  
plt.figure(figsize = (10,3), dpi = 150)
plt.plot(x,y, lw = 0.5)

model.fit(x,y)

y_predict = model.predict(x)


plt.plot(x, y_predict, '-k')
plt.figure(2,(10,3), dpi = 150)

value = (y - y_predict)
print(value.shape)

plt.plot(y_predict, value, lw = 0.5)
plt.hlines(0,-1,31, colors = 'k',lw = 1)

samples = len(y)
dt = 0.05
# obliczenie fft
#print(len(2*abs(fft.rfft(y_predict))/samples))
#print(len(2*abs(fft.rfft(y))/samples))

f_trans = 2*abs(fft.rfft(value, axis = 0)) / samples
print(f_trans.shape)
freq = fft.rfftfreq(len(value), dt)
print(scipy.signal.find_peaks(f_trans[:,0],3))
print(freq[24], freq[33])


fig = plt.figure(figsize=(10,3), dpi=150)
plt.plot(freq, f_trans, 'g')
plt.xlim(0, 5)
plt.xlabel('częstotliwość')
plt.ylabel('amplituda')
plt.vlines(freq[24], 0,f_trans[24])
plt.vlines(freq[33], 0,f_trans[33])
plt.grid()
plt.show()
```

```python
# składowe sygnału
# amplituda
a = array([2.5, 2.3, 4.3])
# częstotliwość
f = array([0.81, 0.84, 1.34]) #0.8
# faza początkowa
fi = array([0.9, 1.2, 3.9])

# ------------------- czas próbkowania ------------------------
t_max = 100
# ------------------- liczba próbek ---------------------------
samples = 500


# In[354]:


# interwał próbkowania
dt = t_max/samples
print(dt)

t = linspace(0, t_max, num = samples)


# In[355]:


y = []
for c in range(0, len(a)):
    y.append(a[c]*sin(2*pi*f[c]*t + fi[c]))


# In[356]:


# wykres składowych sygnału
fig = plt.figure(figsize=(10,3), dpi=150)

for y_i in y:
    plt.plot(t, y_i, '-')
plt.grid()
plt.xlim(0, 20)
plt.xlabel('czas')
plt.ylabel('amplituda')
plt.show()


# In[357]:


# sygnał wypadkowy (suma składowych)
fig = plt.figure(figsize=(10,3), dpi=150)

fy = y[0].copy()
for i in range(1, len(y)):
    fy += y[i]

plt.plot(t, fy, 'r')
plt.xlabel('czas')
plt.ylabel('amplituda')
plt.grid()
plt.show()


# In[358]:


# szum
fy0 = fy.copy()
#print(fy0)

noise = 0
for i in range(1000):
    noise += random.rand(samples)


# In[359]:


fy += 0.3*(noise - mean(noise))
#print(fy)


# In[360]:


# sygnał wypadkowy i sygnał zaszumiony
fig = plt.figure(figsize=(10,3), dpi=150)

plt.plot(t, fy0, 'r')
plt.plot(t, fy, 'b', ms=2, linewidth = 0.5)
plt.legend(['sygnał', 'sygnał + szum'])
plt.xlabel('czas')
plt.ylabel('amplituda')
plt.grid()
plt.show()


# In[361]:

print(len(fy))
# obliczenie fft
f_trans = 2*abs(fft.rfft(fy)) / samples
print(len(f_trans))
freq = fft.rfftfreq(len(fy), dt)
print(len(freq))


# In[362]:


# widmo częstotliwościowe
fig = plt.figure(figsize=(10,3), dpi=150)

plt.plot(freq, f_trans, 'g')

plt.xlim(0, 5)
plt.xlabel('częstotliwość')
plt.ylabel('amplituda')
plt.grid()
plt.show()
```
